## Scraping Artist Names from Wikipedia

### Getting one name

```python
#Scrape one name first
import time
import requests
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_hip_hop_musicians#cite_note-Rap-1'
```

```python
source = requests.get(url).text
```

```python
soup = BeautifulSoup(source,'html')
print(soup.prettify())
```

```python
artist = soup.find_all('div', class_='div-col columns column-width')
```

```python
len(artist)
```

```python
artist[26].find_all('li')[1].text
```

```python
len(artist[26].find_all('li'))
```

### Getting all artists on the wikipedia link

```python
artists = []
for i in range(len(artist)):
    for j in range(len(artist[i].find_all('li'))):
        artists.append(artist[i].find_all('li')[j].text)
```

```python
artists
```

```python
len(artists)
```

```python
#Saving the list with all the artists
import pickle
 
with open('hip-hop-artist.pkl', 'wb') as f:
    pickle.dump(artists, f)
```

### Scraping all artist in ranker

```python
all_artists = []
link_root = 'https://www.ranker.com/crowdranked-list/the-greatest-rappers-of-all-time?page='

for i in tqdm(range(34)):
    all_artists_soup = []
    response = requests.get(link_root+str(i)).text
    bsoup = BeautifulSoup(response,'lxml')
    all_artists_soup = bsoup.find_all('div', class_='listItem__data')
    for j in range(len(all_artists_soup)):
        try:
            all_artists.append(all_artists_soup[j].find('meta', itemprop='name')['content'])
        except:
            pass
```

## Keeping the artists in both lists

```python
def get_common_artists(scrape1, scrape2):
   # Produce the set of *unique* common artists in the two lists.
   return list(set(scrape1) & set(scrape2))
```

```python
#getting the two lists of artist from pickle
import pickle
```

```python
with open('hip-hop-artist.pkl', 'rb') as f:
    artists_wiki = pickle.load(f)
```

```python
with open('hip-hop-artist-ranker.pkl', 'rb') as f:
    artists_ranker = pickle.load(f)
```

```python
print(len(artists_wiki),"and",len(artists_ranker))
```

```python
rap_artists = get_common_artists(artists_wiki,artists_ranker)
```
