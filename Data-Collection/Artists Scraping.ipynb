## Scraping Artist Names

```python
#Scrape one name first
import time
import requests
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_hip_hop_musicians#cite_note-Rap-1'
```

```python
source = requests.get(url).text
```

```python
soup = BeautifulSoup(source,'html')
print(soup.prettify())
```

```python
artist = soup.find_all('div', class_='div-col columns column-width')
```

```python
len(artist)
```

```python
artist[26].find_all('li')[1].text
```

```python
len(artist[26].find_all('li'))
```

```python
artists = []
for i in range(len(artist)):
    for j in range(len(artist[i].find_all('li'))):
        artists.append(artist[i].find_all('li')[j].text)
```

```python
import pickle
 
with open('hip-hop-artist.pkl', 'wb') as f:
    pickle.dump(artists, f)
```
